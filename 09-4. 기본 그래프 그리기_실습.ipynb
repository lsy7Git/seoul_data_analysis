```python
import pandas as pd
```

# 멧플롯립 라이브러리 임포트

```python
import matplotlib.pyplot as plt
```

# x축 y축 데이터 준비

```python
x = ['a','b','c','d','e']
y = [1,3,2,10,7]
```

# 그래프 그리기

## 선그래프

```python
# 시간의 흐름에 따른 값의 변화
plt.plot(x,y)
plt.show()
```

## 막대그래프

```python
# x축 값들의 크기를 비교할 때
plt.bar(x,y)
plt.show()
```

## 가로 막대그래프

```python
plt.barh(x,y)
plt.show()
```

## 산점도

```python
plt.scatter(x,y)
plt.show()
```

## 세가지 그래프를 한번에 그리고 옵션 추가하기

```python
plt.plot(x,y, label='plot')
plt.bar(x,y, label='bar')
plt.scatter(x,y, label='scatter')
plt.title('Test Graph', size=15)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
